```python
import random
```

```python
def play(stay=False):
    car_location = random.randint(0, 2)
    choice = random.randint(0,2)
    random_goat = random.randint(0,1)

    goats = [0,1,2]
    goats.remove(car_location)

    if choice == car_location:
        reveal = goats[random_goat]
    else:
        goats.remove(choice)
        reveal = goats[0]
    
    print "Car at : " + str(car_location)
    print "Original Choice : " + str(choice)
    print "Reveal Goat : " + str(reveal)

    if stay:
        print "Staying!!"
        if choice == car_location:
            print "You Win"
            return True
    else:
        print "Switch choice!!"
        if choice != car_location:
            print "You Win"
            return True
            
    print "You lose"
    return False
```

```python
def simulate(stay, n):
    wins = 0
    for x in xrange(n):
        print "-------------------------"
        print "Simulation: #{}".format(x)
        if play(stay):
            wins+=1
            
    if stay:
        stay_or_switch = "stay"
    else:
        stay_or_switch = "switch"

    print "\n\n\nWinning Chance if you {}: {}%".format(stay_or_switch, wins * 100/n)
```

```python
simulate(False, 1000)
```

```python
simulate(True, 1000)
```
